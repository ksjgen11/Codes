# Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.60, random_state = 0)
plt.scatter(X[:,0], X[:,1], s = 50)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters =4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

centers = kmeans.cluster_centers_

plt.scatter(X[:,0], X[:,1], c= y_kmeans, s = 50, cmap = 'viridis')
plt.scatter(centers[:,0], centers[:,1], c = 'k', s = 200, alpha = 0.5)
```

```python
from sklearn.datasets import load_digits
digits = load_digits()

fig, ax = plt.subplots(3,5)

for i in range(3):
    for j in range(5):
        ax[i][j].axis('off')
        ax[i][j].imshow(digits.data[i*4+j].reshape(8,8), cmap = 'binary',)
```

```python
kmeans = KMeans(n_clusters = 10, random_state = 0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize = (8,3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.axis('off')
    axi.imshow(center, cmap = plt.cm.binary)
```

# Gaussian mixture model

```python
X, y_true = make_blobs(n_samples = 400, centers = 4, cluster_std = 0.60, random_state = 0)
X = X[:,::-1]
plt.scatter(X[:,0], X[:,1], c='gray', alpha = 0.4)
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 0)
clusters = kmeans.fit(X).predict(X)
centers = kmeans.cluster_centers_
plt.scatter(X[:,0], X[:,1], c=clusters, s = 40, cmap = 'viridis')
plt.scatter(centers[:,0], centers[:,1], c='r')
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components = 4).fit(X)
labels = gmm.predict(X)
plt.scatter(X[:,0], X[:,1], c=labels, s = 40, cmap = 'viridis')

probs = gmm.predict_proba(X)
print(probs[:5].round(3))
```

```python
size = 50*probs.max(1) **2
plt.scatter(X[:,0], X[:,1], c=labels, s = size, cmap = 'viridis')
```
