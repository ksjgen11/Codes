# Scatter plot

## multivariate data analysis

- matrix or data frame

- observations == values

### scatter plot
 - 2 variables - variation visualize
 - visualize the relationships

pairs - plotting more than 2 variables

```r
wt <- mtcars$wt
mpg <- mtcars$mpg
plot(wt, mpg,  # plotting scatter plot(x axis, y axis)
    main = 'Car weight - mpg',  # title
    xlab = 'Car weight',  # label of x axis
    ylab = 'Miles Per Gallon', # label of y axis
    col = 'red', # color of points
    pch = 19) # type of points
```

```r
vars <- c('mpg', 'disp', 'drat', 'wt')
target <- mtcars[, vars]
pairs(target, main = 'multi plots')
```

# Correlation analysis

 - using with scatter plot
 - correlation coefficient R
 - R value increases that it is proportional to the line plot
 - -1 < R < 1

```r
beers <- c(5, 2, 9, 8, 3, 7, 3, 5, 5, 5)
bal <- c(0.1, 0.03, 0.19, 0.12, 0.04, 0.095, 0.07, 0.06, 0.06, 0.06)
tbl = data.frame(cbind(beers, bal))
tbl; class(tbl)
plot(bal~beers, data=tbl) #scatter plot
res = lm(bal~beers, data=tbl) # regression formula
abline(res) # plotting regression line
cor(beers, bal) # analyzing correlations
```

```r
cor(iris[,1:4])
```

```r
income <- c(125, 100, 40, 35, 41, 29, 35, 24, 50, 60) * 1000
edu <- c(19, 20, 16, 16, 18, 12, 14, 12, 16, 17)
table <- data.frame(cbind(income, edu))
table
plot(edu~income, data = table)
res = lm(edu~income, data = table)
abline(res)
cor(income, edu)
```

```r
GPA <- c(3.1, 2.4, 2.0, 3.8, 2.2, 3.4, 2.9, 3.2, 2.7, 2.5)
TV <- c(14, 10, 20, 7, 25, 9, 15, 13, 4, 21)
table <- data.frame(cbind(TV, GPA))
table
plot(GPA~TV, data = table)
res = lm(GPA~TV, data = table)
abline(res)
cor(TV, GPA)
```

# line plot

 - when one variable is time
 

```r
month <- 1:12
late1 <- c(5, 8, 7, 9, 4, 6, 12, 13, 8, 6, 6, 4)
late2 <- c(4, 6, 5, 8, 7, 8, 10, 11, 6, 5, 7, 3)
plot(month,  # x axis -> usually time points are x axis
     late1,  # y data
     main = 'Late students', 
    type = 'l', # the type of graph
    lty = 1, # type of line
    lwd = 1, # width of line
    xlab = 'Month', ylab = 'Late count', 
    col = 'red')
lines(month, # graph overlapped previous graph
     late2, 
     type = 'l', 
     col = 'blue')
```

```r
year <- 2015:2026
pop <- c(51014, 51245, 51446, 51635, 51811, 51973, 32123, 52261, 52338, 52504, 52609, 52704)

plot(year, pop, type = 'l', lty = 1, lwd = 1)
```

```r
quat <- c(20144, 20151, 20152, 20153, 20154, 20161, 20162, 20163, 20164, 20171, 20172, 20173)
man <- c(73.9, 73.1, 74.4, 74.2, 73.5, 73.0, 74.2, 74.5, 73.8, 73.1, 74.5, 74.2)
woman <- c(51.4, 50.5, 52.4, 52.4, 51.9, 50.9, 52.6, 52.7, 52.2, 51.5, 53.2, 53.1)

plot(quat, man, type = 'l', col = 'red', ylim = c(50, 75) )
lines(quat, woman, type = 'b', col = 'blue')
```
