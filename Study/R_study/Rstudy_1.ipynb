## Factor
 - factor is string variable, similar with vector
 - only specific values are allowed (A, B, AB,O blood types)

```r
blood.type <- factor(c('A', 'A', 'AB', 'O', 'B'))
blood.type
```

```r
blood.type[6] <- 'D'
blood.type
```

```r
as.numeric(blood.type)
```
