## 4.2. 바이트에 대한 기본 지식

이진 시퀀스를 위해 사용되는 내장 자료형 : bytes, bytearry
*bytes : 불변형,  bytearray : 가변형*
bytes, bytearray : 0~255 사이의 정수가 들어있음(정수형 시퀀스) >> 이진 시퀀스 슬라이싱 >. 동일한 자료형의 이진 시퀀스 만들어짐.

이진시퀀스 리터럴 표기법 >> 실제로 아스키 텍스트 경우 많음 >> 다음 3가지 형태로 출력
1. 화면에 출력 가능한 아스키 문자
2. 탭, 개행문자,캐리지리턴,백슬래시 >> 이스케이프 시퀀스로 출력(|t, |n, |r, ||)
3. 그 외의 값 >> 16진수 이스케이프 시퀀스로 출력

bytes, bytearray는  format(), format_map() 제외한 str 제공 메서드 모두 지원. 유니코드 데이터에 관련된 메서드 지원.

bytes, bytearray 객체 생성 인수
1. str, encoding 키워드 인수
2. 0~255 사이의 값 제공하는 반복 가능형
3. 정수 인수 >> 인수의 개수만큼 널 바이트로 초기화된 이진 시퀀스 생성
4. 버퍼 프로토콜을 구현하는 객체 >> 원본 객체 바이트 복사 후 바이트 시퀀스 새로 생성

## 4.6. 유니코드 정규화

유니코드 : 결합문자 >> 문자열 비교 어려움 >> 유니코드 정규화 >> 문자열 비교

NFC (Normalization Form C) : 코드 포인트 조합 >> 가장 짧은 동일 문자열 생성
*기호 옴>-NFC->그리스어 대문자 오메가*
NFD : 조합된 문자 >> 기본 문자, 결합문자로 분리

NFKC, NFKD : K는 호환성을 나타냄. 정규화의 더 강력한 형태. 호환성 문자는 포매팅 손실이 발생하더라도ㅗ 선호하는 형태의 호환성 분활로 치환 됨.

### 4.6.2. 정규화된 텍스트 매칭을 위한 유틸리티 함수

### 4.6.3. 발음 구별 기호 제거



```python
#결합표시 기호 제거

import unicodedata
import string

def shave_marks(txt):
    """발음 구별 기호를 모두 제거한다."""
    norm_txt = unicodedata.normalize('NFD',txt) #기본문자, 결합표시로 분해
    shaved = ''.join(c for c in norm_txt
                    if not unicodedata.combining(c)) #결합표시 걸러냄
    return unicodedata.normalize('NFC',shaved) #문자 모두 재결합시킴
```

### 4.9.2. os 모듈함수에서 str, bytes

GNU/리눅스 커널은 유니코드 모름. >> os 파일명은 인코딩 체계에 맞지 않는 바이트 시퀀스로 구성 >> str로 디코딩 불가능 >> 파일명, 경로명 받는 os 모듈함수는 str, bytes형 인수 받음 >> str 인수로 함수 호출>> sys.getfilesystemcoding() 의해 지정된 코덱 통해 변환 >> 운영체계 응답 동일 코덱 통해 디코딩
*위 과정은 유니코드 샌드위치 모델에 따라 작동*

처리할 수 없는 파일명 이용/수정 >> bytes 인수 os 함수에 전달 >> bytes 반환값
*파일명이나 경로명에 깨진 문자 얼마나 많은지 상관없이 이 방식 쓸 수 o*

fsencode(파일명) : 파일명 str >> sys.getfilesystemcoding()이 반환한 코덱명 이용 >> bytes 로 인코딩.
                   파일명 str >> 그대로 반환

fsdecode(파일명) : 파일명 bytes >> sys.getfilesystemencoding()이 반환한 코덱명 이용 >> str로 디코딩
                   파일명 str >> 그대로 반환
