# Chapter 6 - 일급 함수 디자인 패턴

패턴에 참여하는 일부 클래스의 객체를 간단한 함수로 교체하면 획일적으로 반복되는 코드의 상당 부분을 줄일 수 있다.<br>

## 전략 패턴

![](https://i.ibb.co/X48R7zG/image.png)<br>
전략 패턴은 위 그림과 같은 형태를 따른다.

- 콘텍스트: 일부 계산을 서로 다른 알고리즘을 구현하는 교환 가능한 컴포넌트에 위임함으로써 서비스를 제공한다.
- 전략: 여러 알고리즘을 구현하는 컴포넌트에 공통된 인터페이스.
- 구체적인 전략: 전략의 구성 컴포넌트 중 하나.

『디자인 패턴』에서는 전략 패턴을 **"일련의 알고리즘을 정의하고 각각을 하나의 클래스 안에 넣어서 교체하기 쉽게 만든다. 전략을 이용하면 사용하는 클라이언트에 따라 알고리즘을 독립적으로 변경할 수 있다."** 라고 설명한다.<br>
여기까지 고전적인 전략 패턴이고, 이를 함수지향 전략을 이용하여 구현하면 좀 더 간단해진다.<br>
구제척인 전략을 클래스로 구현하는 것이 아니라 단순한 함수로 구현하는 것이다.<br>
함수는 사용자 정의 클래스보다 훨씬 가볍고 파이썬이 모듈을 컴파일할 때 단 한 번만 생성되므로 플라이웨이트가 필요하지 않다.

## 명령 패턴

![](https://i.ibb.co/X4tdQDb/123.png)<br>
명령 패턴은 위 그리과 같은 형태를 따른다.<br>
명령 패턴의 목적은 연산을 실행하는 객체(호출자)와 연산을 구현하는 객체(수신자)를 분리하는 것이다.<br>
기본 개념은 명령 객체를 수신자와 호출자 사이에 놓고, 명령은 execute()라는 단 하나의 메서드로 인터페이스를 구현한다. execute()는 원하는 연산을 수행하기 위해 수신자가 가지고 있는 메서드를 호출한다. 이런 방식을 사용하면 호출자는 수신자의 인터페이스를 알 필요가 없으며, 명령의 서브클래스를 통해 서로 다른 수신자를 추가할 수 있다.<br><br>

여기서 호출자에 객체 대신 간단히 함수로 바로 지정할 수 있다. command.execute()를 호출하는 대신, 호출자는 단지 command()를 호출하면 된다. MacroCommand는 \_\_call\_\_() 메서드를 가진 클래스로 구현할 수 있다. MacroCommand의 객체는 콜러블이 되며 향후에 호출할 함수들의 리스트를 가진다.
