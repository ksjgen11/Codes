# 13. 연산자 오버로딩

연산자 오버로딩 >> 사용자 정의 객체가 중위연상자(+,|), 단항 연산자(-,~) 사용 가능

## 13.1. 연산자 오버로딩 기본 지식

내장 자료형에 대한 연산자는 오버로딩할 수 없다.
새로운 연산자를 생성할 수 없으며, 기존 연산자를 오버로딩만 할 수 있다.
is, and, or, not 연산자 오버로딩할 수 없다.
&,|,~ 비트 연산자는 오버로딩 가능.
  >>융통성, 사용성, 안정성 유지

## 13.2. 단항 연산자

단항 연산자(연산자와 연관된 특별 메서드)
 -(__neg__) : 단항 산술 부정. x=-2 >> -x=2
 +(__pos__) : 단항 산술 덧셈. 일반적으로 x==+x
 ~(__invert__) : 정수형의 비트 반전. ~x == -(x+1)
 
abs()도 단항 연산자로 나열한다.

self 인수 하나를 받는 특별 메서드 구현 >> 단항 연산자 구현
  언제나 새로운 객체를 반환해야 한다.(연산자 핵심규칙) >> self 수정x, 적절한 자료형의 객체 새로 생성해 반환해야.
  
  + : slef의 사본 반환이 좋음. abs() >>스칼라형 숫자
  ~ : 객체 관계 매핑에서 사용 >> SQL의 WHERE 구를 부정 결과 반환

x==+x가 성립하지 않을 때도 있음.

## 13.3. + 오버로딩 >> 벡터 더하기

짧은 벡터 빈 공간 0으로 채워서 더하기.

파이썬 a+b 처리 절차
 1. a 에 __add__() 정의 >> a.__add__(b) 호출 >> 결과 NotImplemented 아님 반환
 2. a에 __add__() 정의x/NotImplemented 반환 >> b에 __radd__() 정의되어 있는지 확인 >> b.__add__(a) 호출 >> 결과 NotImplemented 아님 반환
 3. b에 __radd__() 정의 x/NotImplemented 반환 >> '지원하지 않는 피연산자형', typeerror
 
__radd__() : __add__()의 역순 버전

## 13.4. * 오버로딩 >> 벡터 스칼라 곱

스칼라곱 : 두 벡터의 내적 >> numpy.dot() 이용
* 연산자 >> 벡터의 각 항목에 값을 곱함.

구스 타이핑 >> 명시적 방법 사용


중위 연산자 메서드 명
  연산자  정방향  역순   인플레이스  설명
  + ; __add__() ; __radd__() ; __iadd__() ; 덧셈/연결
  - ; __sub__() ; __rsub__() ; __isub__() ; 뺄셈

  / ; __truediv__() ; __rtruediv__() ; __itruediv__() ; 참 나눗셈
  // ; __floordiv__() ; __rfloordiv__() ; __ifloordiv__() ; 플로어 나눗셈
  % ; __mod__() ; __rmod__() ; __imod__() ; 모듈로 나머지 연산
  divmod() ; __divmod__() ; __rdivmod__() ; __idivmod__() ; 플로어 나눗셈의 몫, 나머지 튜플로 반환
  **, pow() ; __pow__() ; __rpow__() ; __ipow__() ; 누승
  @ ; __matmul__() ; __rmatmul__() ; __imatmul__() ; 행렬 곱셈
  & ; __and__() ; __rand__() ; __iand__() ; 비트단위 곱
  | ; __or__() ; __rorv_() ; __ior__() ; 비트단위 합
  ^ ; __xor__() ; __rxor__() ; __ixor__() ; 비트단위 배타합
  << ; __lshift__() ; __rlshift__() ; __ilshift__() ; 비트단위 왼쪽 시프트
  >> ; __rshift__() ; __rrshift__() ; __irshift__() ; 비트단위 오르쪽 시프트
  

## 13.5. 향상된 비교 연산자

비교 연산자 : ==, !=, >, < , >=, <=
정방향, 역순 연산자에 동일한 세트의 메서드 사용됨
ㅇ ==, != 연산자는 역순 메서드 실패 >> typeerror 대신 객체 ID 비교

피연산자 평가 시 지나친 자유분방함은 예기치 못한 결과 낳을 수 있음.
 >> 모호함에 직면할 때는 추측하려는 유혹을 거부하라

## 13.6. 복합 할당 연산자

인플레이스 연산자 메서드 정의 >> 연산자 새로운 객체 생성하지 않고 피연산자를 직접 변경

+= 연산자가 +보다 자유롭다 ; +는 양쪽 피연산자가 동일한 자료형이어야, +=는 상관 없음.

__add__() : AddableBingoCage() 호출해 생성된 새로운 객체 반환
__iadd__() : 객체 자신을 변경 후 self 반환
