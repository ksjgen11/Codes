## 14.2. 반복형과 반복자

반복형 : iter() 내장 함수가 반복자를 가져올 수 있는 모든 객체와 반복자를 반환하는 __iter__() 메서드를 구현하는 객체는 반복형이다. 0에서 시작하는 인덱스를 받는 __getitem__() 메서드를 구현하는 객체인 시퀀스도 마찬가지다.

반복자에 대한 표준 인터페이스 >> __next__(), __iter__() 정의 함.

__next__() : 다음에 사용할 항목을 반환. 항목 남아 있지 않으면 StopIteration 발생.
*StopIteration : 반복자 모두 소진됨을 알려줌*
__iter__() : self를 반환 >> for 루프 등 반복형 필요한 곳에 반복자 사용 가능하게 함. 

반복자 : 다음 항목을 반환하거나, 다음 항목이 없을 때 StopIteration 예외를 발생시키는 인수를 받지 않는  __next__() 메서드를 구현하는 객체. 파이썬 반복자는 __iter__() 도 구현하므로 반복형이기도 함.

반복자 : 개별 항목 반환하는 __next__(), self 반환하는 __iter__() 가짐.

반복자는 반복형임, 반복형은 반복자 아님.

다중 반복 지원하려면 >> 동일한 반복형 객체로부터 여러 독립적인 반복자 가질 수 있어야, 각 반복자는 고유한 내부 상태 유지해야 : iter(my_iterable) 호출마다 독립적인 반복자 새로 만들어져야 >> SentenceIteration 클래스 필요.
