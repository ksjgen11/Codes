# 11. 인터페이스

## 11.1 파이썬 문화에서 인터페이스, 프로토콜

프로토콜 : 동적 자료형 제공하는 언어(ex 파이썬)에서 다형성을 제공하는 비공식 인터페이스

파이썬 : interface라는 키워드 없음. 모든 클래스는 인터페이스 갖고 있음.  
         인터페이스:클래스 상속하거나 구현한 공개 속성(메서드, 데이터 속성,특별메서드)들의 집합
         보호된 속성, 비공개 속성은 인터페이스에 속하지 않음.

공개 데이터 속성을 인터페이스로 사용 : <객체>.<속성> 구문으로 데이터 속텅을 프로퍼티로 변환 가능

예11-1)x,y 읽기 전용 프로퍼티로 변경. but 인터페이스의 핵심부분은 바뀌지 않음.

인터페이스 : 시스템에서 어떤 역할을 할 수 있게 해주는 객체의 공개 메서드의 일부
   *프로토콜이라고 불렸음. 프로토콜은 상속과 무관. 프로토콜은 인터페이스이지만 비공식적임.*
             클래스 : 여러 프로토콜 구현 > 객체가 여러 역할 하게 함

X와 같은 객체, X 프로토콜, X 인터페이스 : 다 같은 말임

## 11.6. 표준 라이브러리의 ABC

표준 라이브러리에 ABC 포함되어 있음.
대부분의 ABC는 collections.abc 모듈에 정의 되어 있음.

### 11.6.1. collections.abc의 ABC

믹스인 메서드 : 클래스를 상속하지 않고 사용할 수 있는 메서드

ABC에서는 일반적으로 다중 상속이 문제되지 않음.

Iterable, Container, Sized : 모든 컬렉션 이 ABC 상속하거나 호환되는 포토콜 구현해야 함. 각각 __iter__() 통해 반복, __contains__() 통해 in 연산자, __len__() 통해 len() 지원함

Sequence, Mapping, Set : 불변 컬렉션형, 가변형 서브클래스가 있음.

MappingView :  items(), keys(), values() 메서드에서 반환된 객체는 각기 ItemsView, KeysView, ValueView를 상속. ItemsView, ValuesView는 Set를 상속

Callable, Hashable : colletions.abc가 파이썬 표준 라이브러리안에서 ABc를 정의한 최초의 패키지 임. ABC 컬렉션과 밀접한 연관이 있지는 않음. isinstance()와 함께 사용됨.

Iteratoar : Iterable 상속

# 12. 내장 자료형 상속과 다중 상속

## 12.1. 내장 자료형의 상속은 까다롭다

공식적으로 CPython은 내장 자료형의 서브클래스에서 오버라이드한 메서드가 언제 호출되는지, 혹은 호출되지 않는지에 대해 명확한 규칙을 정의하지 않는다. 일반적으로 서브클래스에서 오버라이드한 메서드는 같은 객체의 다른 내장 메서드에 의해 결코 호출되지 않는다.

슈퍼클래스에서 구현된 메서드 안에서 호출하더라도 메서드 검색은 대상 객체의 클래스에서 시작해야 한다. (객체지향 프로그래밍의 기본 규칙)
*내장 자료형은 이를 어김,, 내장 메서드가 호출하는 다른 클래스의 오버라이드된 메서드인 경우도 어김*

## 12.2. 다중 상속과 메서드 결정 순서

메서드 결정 순서 (MRO) 
    __mro__ 속성은 현재 클래스부터 object 클래스까지 슈퍼클래스들의 MRO를 튜플 형태로 저장함.
    >> 다중 상속 시 메서드 순서 결정됨

객체 메서드를 클래스에 직접 호출할 떄는 self를 반드시 명시해야 함. : 바인딩되지 않은 메서드에 접근하는 것이기 때문

MRO는 상속 그래프, 서브클래스 정의에 나열된 슈퍼클래스들의 순서 고려

## 12.3. 실세계에서의 다중 상속

파이썬;  표준 라이브러리인 collection.abc 패키지에서 다중 상속 많이 사용
자바 ; 인터페이스 다중 상속 지원

## 12.4. 다중 상속 다루기

상속은 다양한 목적으로 사용됨.
다중 상속은 대체 수단, 복잡도 증가 >> 이해 어렵고 불안정하게 설계하기 쉬움.

뒤엉킨 클래스 그래프 만들어지는 것 예방법
1. 인터페이스 상속과 구현 상속 구분 : 상속 이유 명확히 하기
        인터페이스 상속은 'is-a' 관계 의미하는 서브타입 생성
        구현 상속은 재사용을 통해 코드 중복 피함
        
2. ABC를 이용해 인터페이스 명확히 하기
    최근 파이썬에서 ABC : 인터페이스를 정의하기 위해 만들어진 클래스
    3.4 버전~  : ABC로 명시하기 위해 abc.ABC나 다른 ABC 상속하는 것
    
3. 코드 재상용 위해 믹스인 사용
    구현 상속 클래스는 믹스인 클래스로 만들어야
    믹스인 클래스 : 새로운 자료형 정의하지 않음, 재사용할 메서드를 묶어 놓을 뿐
    믹스인 클래스로 객체 생성하면 안됨
    믹스인 클래스 상속하는 구상 클래스는 다른 클래스도 상속해야
    각각의 믹스인 클래스는 밀접하게 연관된 메서드 구현해 구체적 행위 제공해야
    
4. 이름을 통해 믹스인임을 명확히 하기
    클래스명 뒤에 Mixin 붙이기

5. ABC->믹스인 가능 but, 믹스인=ABC 아님
    ABC는 자료형 정의하지만 믹스인은 자료형 정의하지 않음.
    ABC는 다른 클래스의 유일한 기저 클래스 될 수 있지만, 믹스인 하나만 이용해 서브 클래스 정의하면 안됨.
    ABC에서 구현된 구상메서드는 해당 ABC나 슈퍼클래스의 메서드만 사용 가능

6. 2개 이상의 구상 클래스에서 상속받지 않게
    구상클래스는 0~1개의 구상 슈퍼클래스 가져야
    나머지는 ABC나 믹스인이어야
    
7. 사용자에게 집합 클래스 제공
    집합 클래스 : ABC나 믹스인 조합해 호출 코드에 유용한 기능을 제공할 수 있을 때 이를 통합하는 클래스
    집합 클래스 제공 >> 믹스인 기억할 필요 없음, class문 안에서 특정 순서대로 선언해야하는지 확인할 필요 없음.
    
8. 클래스 상속보다 객체 구성 사용
    상속 >> 계층구조로 깔끔하게 정리해 보기 쉬움
    구성 >> 융통성 향상
    상속 트리의 높이가 높아지면 구조 불안정해 짐.
