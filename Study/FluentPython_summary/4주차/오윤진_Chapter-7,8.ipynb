## 8.1. 변수

변수는 객체에 붙은 레이블임. 상자가 아님.

## 8.2. 정체성, 동질성, 별명

별명 : 두 변수가 동일 객체에 바인딩 되어야 함.
       그 두 변수가 a, b라고 하면 a==b이다. 하지만 a,b의 정체성은 다름.

정체성 : 생성 후에 결코 변경되지 않음.
         메모리 내의 객체 주소
         is 연산자 통해 정체성 비교 가능
         id() : 정체성을 나타내는 정수를 반환
                cPython : 객체의 메모리 주소 반환
                다른 파이썬 인터프리터 : 메모리 주소가 아닌 다른 값 반환할 수도

### 8.2.1. == , is

== : 객체의 값 비교
is : 객체의 정체성 비교

변수 vs singleton : is 사용해야.

x is None
x is not None

is가 ==보다 빠름.

a==b : a.__eq__(b)의 편리구문
       object 객체의 __eq__() 메서드 : 객체의 id 비교 >> is와 같은 결과
       대부분의 내장 자료형 : __eq__()메서드를 오버라이드 해 객체 값 비교

### 8.2.2. 튜플의 상대적 불변성

튜플  : 불변형. but 참조된 항목 가변형>>참조 항목은 변할 수 있음.
