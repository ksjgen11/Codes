```r
#Day 1_part 2
data = read.table('advantcode_day1_data.txt')

for(i in data[,1]){
    for (j in data[,1]){
        for (k in data[,1]){
            if (i + j + k == 2020){
            print(c(i, j, k, i*j*k))
            }
                
        }
    }
}
```

```r
#day 2_part 1
setwd('/home/ksjgen11/codingproject/study/workplace')

file1 <- read.table('advantcode_day2_data.txt', sep = '-')


file2 <- data.frame(do.call('rbind', strsplit(as.character(file1$V2), split = ' ', fixed= TRUE)))

file3 <- data.frame(do.call('rbind', strsplit(as.character(file2$X2), split = ':', fixed= TRUE)))

file <- data.frame(file1$V1, file2$X1, file3, file2$X3)
names(file) <- c('min', 'max', 'sign', 'pwd')
file
```

```r
library(stringr)
counts <- str_count(file$pwd, file$sign)
counts
filex <- data.frame(file, counts)
filex
```
